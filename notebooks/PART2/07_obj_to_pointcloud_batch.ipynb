# 07 — OBJ to Point Cloud Batch Export

This notebook converts the generated Alice-related OBJ models into standardised point cloud CSV files for TouchDesigner.

Each OBJ model is sampled into a fixed number of points. The output format is kept consistent as:

- `x`
- `y`
- `z`

This structure is required for the TouchDesigner point cloud workflow.

```python
import os
import numpy as np
import trimesh
import pandas as pd

# =========================
# 参数设置
# =========================
INPUT_DIR = "../../data/raw/alice_models"
OUTPUT_DIR = "../../data/processed/alice_pointclouds"
N_POINTS = 5000

os.makedirs(OUTPUT_DIR, exist_ok=True)

# =========================
# 处理函数（不改）
# =========================
def mesh_to_pointcloud(mesh_path, n_points=5000):
    mesh = trimesh.load(mesh_path)

    if isinstance(mesh, trimesh.Scene):
        mesh = mesh.dump(concatenate=True)

    points, _ = trimesh.sample.sample_surface(mesh, n_points)
    return points

# =========================
# 自动获取所有 OBJ 文件
# =========================
files = [f for f in os.listdir(INPUT_DIR) if f.endswith(".obj")]

# 可选：排序（建议加，保证顺序稳定）
files.sort()

# =========================
# 批处理
# =========================
for file in files:
    input_path = os.path.join(INPUT_DIR, file)

    print(f"Processing: {file}")

    pts = mesh_to_pointcloud(input_path, N_POINTS)

    df = pd.DataFrame(pts, columns=["x", "y", "z"])

    output_name = file.replace(".obj", ".csv")
    output_path = os.path.join(OUTPUT_DIR, output_name)

    df.to_csv(output_path, index=False)

    print(f"Saved: {output_path}")

print("Done!")
```
